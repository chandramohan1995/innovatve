```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\kcreddy\\Downloads\\Global Superstore.csv')
```

```python
df.drop([12887,14843,7399,329,290] ,axis = 0,inplace = True)
```

```python
df1 = df.drop(df.iloc[:,[0,1,2,3,6,11]],axis =1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df
```

```python
df_num = df1.select_dtypes(include = [np.number])
df_cat = df1.select_dtypes(exclude = [np.number])
```

```python
df_cat_le = df_cat.apply(le.fit_transform)
```

```python
dff  =pd.concat([df_num,df_cat_le],axis =1)
```

```python
dff_corr = dff.corr()
```

```python
dff_corr[(dff_corr['Sales']<0.3)].index.values
```

```python
cols = []
cols.extend(dff_corr[(dff_corr['Sales']<0.3)].index.values)
cols.extend(dff_corr[(dff_corr['Sales']>0.3)].index.values)
```

```python
data =dff[cols]
```

```python
data.isnull().sum()
```

```python
X = data.drop('Sales',axis =1)
X2 = data.drop('Profit',axis =1)
X3 = data.drop ('Discount',axis =1)
```

```python
y = data['Sales']
y2 = data['Profit']
y3 = data['Discount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train ,y_test = train_test_split( X, y ,test_size =0.3)
x2_train ,x2_test,y2_train ,y2_test = train_test_split( X2, y2 ,test_size =0.3)
x3_train ,x3_test,y3_train ,y3_test = train_test_split( X3, y3 ,test_size =0.3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
rf = RandomForestRegressor(n_estimators=500,)
```

```python
gd = GradientBoostingRegressor(n_estimators=1000,loss ='ls')
```

```python
gd2 = GradientBoostingRegressor(n_estimators=1500,loss ='lad')
```

```python
gd_m1 = gd.fit(x_train,y_train)
```

```python
gd_m2 = gd2.fit(x2_train,y2_train)
```

```python
gd_m3 = gd.fit(x3_train,y3_train)
```

```python
print(gd_m1.score(x_train,y_train))
y_pred = gd_m1.predict(x_test)
r2_score(y_pred,y_test)
```

```python
print(gd_m2.score(x2_train,y2_train))
y_pred2 = gd_m2.predict(x2_test)
r2_score(y_pred2,y2_test)
```

```python
print(gd_m3.score(x3_train,y3_train))
y_pred3 = gd_m3.predict(x3_test)
r2_score(y_pred3,y3_test)
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=6)
```

```python
for train, test in kf.split(data):
        train_data = data.iloc[train,:]
        test_data = data.iloc[test,:]
        x_train = train_data.drop(["Sales"],axis=1)
        y_train = train_data["Sales"]
        x_test = test_data.drop(["Sales"],axis=1)
        y_test = test_data["Sales"]
        algo_model = rf.fit(x_train,y_train)
        print(algo_model.score(x_train,y_train))
        y_pred = algo_model.predict(x_test)
        print(r2_score(y_test,y_pred))
```

```python
for train, test in kf.split(data):
        train_data = data.iloc[train,:]
        test_data = data.iloc[test,:]
        x_train = train_data.drop(["Profit"],axis=1)
        y_train = train_data["Profit"]
        x_test = test_data.drop(["Profit"],axis=1)
        y_test = test_data["Profit"]
        algo_model = rf.fit(x_train,y_train)
        print(algo_model.score(x_train,y_train))
        y_pred = algo_model.predict(x_test)
        print(r2_score(y_test,y_pred))
```
